```python
# import the libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import preprocessing, metrics
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
sns.set(style="white", color_codes=True)
import warnings
warnings.filterwarnings("ignore")
```

```python
#Question-1: Principal Component Analysis
df = pd.read_csv(r"CC GENERAL.csv")
```

```python
df.head()
```

```python
df.isnull().any()
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().any()
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
print(x.shape,y.shape)
```

```python
# a. Apply PCA on CC dataset.
#Datasets can be analyzed with PCA so that redundant features can be removed without losing too much information.
pca = PCA(3)   #Instantiate PCA
x_pca = pca.fit_transform(x)
principalDf = pd.DataFrame(data = x_pca, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
finalDf = pd.concat([principalDf, df.iloc[:,-1]], axis = 1)
finalDf.head()
```

```python
# b. Apply k-means algorithm on the PCA result and report your observation if the silhouette score has improved or not?

X = finalDf.iloc[:,0:-1]
y = finalDf.iloc[:,-1]
```

```python
nclusters = 3 # this is the k in kmeans
km = KMeans(n_clusters=nclusters)
km.fit(X)

# predict the cluster for each data point
y_cluster_kmeans = km.predict(X)

# Summary of the predictions made by the classifier
print(classification_report(y, y_cluster_kmeans, zero_division=1))
print(confusion_matrix(y, y_cluster_kmeans))

#finding the accuracy
train_accuracy = accuracy_score(y, y_cluster_kmeans)
print("\nAccuracy for our Training dataset with PCA:", train_accuracy)

#Calculating sihouette Score
score = metrics.silhouette_score(X, y_cluster_kmeans)
print("Sihouette Score: ",score)   #ranges from -1 to +1, high value shows that it is matched more
```

```python
#Question-4: Briefly identify the difference between PCA and LDA
PCA (Principal Component Analysis) is a linear dimensionality reduction technique that uses 
Singular Value Decomposition to project data into a reduced dimensional space. 
Prior to using the SVD, the input data is scaled but not centered for each feature. 
      It employs either a randomized shortened SVD or the complete SVD implementation as implemented by LAPACK.
LDA (Linear Discriminant Analysis) is a classifier with a linear decision boundary that is created by fitting 
class conditional densities to the data and applying Bayes' rule. If all classes have the same covariance matrix, 
the model fits a Gaussian density to each class.
Both LDA and PCA are linear transformation methods, although LDA is supervised whereas PCA is unsupervised 
and disregards class labels. In contrast to PCA, LDA looks for a feature subspace that maximizes class separability.
```
